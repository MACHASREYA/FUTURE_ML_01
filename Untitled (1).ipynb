```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('Sample - Superstore.csv.zip', encoding='latin-1')
print(df.shape)
print(df.dtypes)
print(df.head())
print(df.isnull().sum())
```

```python
# Parse dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Drop duplicates
df.drop_duplicates(inplace=True)

# Select relevant columns
df = df[['Order Date', 'Sales', 'Category', 'Region', 'Quantity', 'Profit']]

# Sort by date
df.sort_values('Order Date', inplace=True)

print(df.head())
```

```python
# Fix: Convert Order Date to datetime FIRST
df['Order Date'] = pd.to_datetime(df['Order Date'])
print(df['Order Date'].dtype)  # should print: datetime64[ns]
```

```python
# Aggregate daily sales
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
daily_sales.columns = ['Date', 'Sales']

# Fix: ensure Date column is datetime
daily_sales['Date'] = pd.to_datetime(daily_sales['Date'])

# Resample to monthly
monthly_sales = daily_sales.set_index('Date').resample('ME').sum().reset_index()

# Create time-based features
monthly_sales['Month'] = monthly_sales['Date'].dt.month
monthly_sales['Year'] = monthly_sales['Date'].dt.year
monthly_sales['Quarter'] = monthly_sales['Date'].dt.quarter
monthly_sales['DayOfYear'] = monthly_sales['Date'].dt.dayofyear
monthly_sales['TimeIndex'] = range(len(monthly_sales))

# Lag features
monthly_sales['Lag_1'] = monthly_sales['Sales'].shift(1)
monthly_sales['Lag_2'] = monthly_sales['Sales'].shift(2)
monthly_sales['Rolling_3'] = monthly_sales['Sales'].shift(1).rolling(3).mean()

# Drop NaN rows
monthly_sales.dropna(inplace=True)
monthly_sales.reset_index(drop=True, inplace=True)

print(monthly_sales.head())
print(f"\nTotal months: {len(monthly_sales)}")
```

```python
plt.figure(figsize=(14, 5))
plt.plot(monthly_sales['Date'], monthly_sales['Sales'], marker='o', color='steelblue', linewidth=2)
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('monthly_sales_trend.png', dpi=150)
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Features and target
features = ['TimeIndex', 'Month', 'Quarter', 'Year', 'Lag_1', 'Lag_2', 'Rolling_3']
X = monthly_sales[features]
y = monthly_sales['Sales']

# Train/test split (keep last 6 months for testing)
split = len(X) - 6
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# Train model
model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.05, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Features and target
features = ['TimeIndex', 'Month', 'Quarter', 'Year', 'Lag_1', 'Lag_2', 'Rolling_3']
X = monthly_sales[features]
y = monthly_sales['Sales']

# Train/test split (keep last 6 months for testing)
split = len(X) - 6
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# Train model
model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.05, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)
```

```python


mae  = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

print(f"MAE  : ${mae:,.2f}")
print(f"RMSE : ${rmse:,.2f}")
print(f"RÂ²   : {r2:.4f}")
```

```python
# Build future rows
last_row = monthly_sales.iloc[-1]
future_rows = []

for i in range(1, 7):
    future_date = last_row['Date'] + pd.DateOffset(months=i)
    future_time_index = last_row['TimeIndex'] + i
    future_month = future_date.month
    future_quarter = future_date.quarter
    future_year = future_date.year

    # Use last known lag values (simplified)
    lag1 = monthly_sales['Sales'].iloc[-1] if i == 1 else future_rows[-1]['Predicted_Sales']
    lag2 = monthly_sales['Sales'].iloc[-2] if i == 1 else (monthly_sales['Sales'].iloc[-1] if i == 2 else future_rows[-2]['Predicted_Sales'])
    rolling3 = monthly_sales['Sales'].iloc[-3:].mean()

    row = {
        'Date': future_date,
        'TimeIndex': future_time_index,
        'Month': future_month,
        'Quarter': future_quarter,
        'Year': future_year,
        'Lag_1': lag1,
        'Lag_2': lag2,
        'Rolling_3': rolling3
    }
    pred = model.predict(pd.DataFrame([row])[features])[0]
    row['Predicted_Sales'] = pred
    future_rows.append(row)

future_df = pd.DataFrame(future_rows)
print(future_df[['Date', 'Predicted_Sales']])
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

# Historical
ax.plot(monthly_sales['Date'], monthly_sales['Sales'],
        label='Historical Sales', color='steelblue', linewidth=2, marker='o', markersize=4)

# Actual vs Predicted (test period)
test_dates = monthly_sales['Date'].iloc[split:]
ax.plot(test_dates, y_pred, label='Model on Test Set', color='orange',
        linewidth=2, linestyle='--', marker='s', markersize=5)

# Future forecast
ax.plot(future_df['Date'], future_df['Predicted_Sales'],
        label='6-Month Forecast', color='green', linewidth=2.5,
        linestyle='--', marker='^', markersize=7)

# Shaded forecast region
ax.axvspan(future_df['Date'].iloc[0], future_df['Date'].iloc[-1],
           alpha=0.08, color='green', label='Forecast Period')

ax.set_title('Sales Forecast â Next 6 Months', fontsize=17, fontweight='bold')
ax.set_xlabel('Date', fontsize=13)
ax.set_ylabel('Monthly Sales ($)', fontsize=13)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('sales_forecast.png', dpi=150)
plt.show()
```
